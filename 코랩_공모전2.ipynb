```python
import json

# JSON 파일 경로 (필터링된 데이터셋을 저장한 경로)
json_file_path = "/content/drive/MyDrive/filtered_data.json"

# JSON 파일 불러오기
with open(json_file_path, 'r', encoding='utf-8') as f:
    filtered_data = json.load(f)

# filtered_data 확인
print(f"filtered_data 크기: {len(filtered_data)}")
print(f"첫 번째 데이터 예시: {filtered_data[0]}")
```

```python
from sentence_transformers import SentenceTransformer, util
import json

# Sentence-BERT 모델 로드
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# 두 텍스트 간의 코사인 유사도 계산
def cosine_similarity(embeddings1, embeddings2):
    return util.pytorch_cos_sim(embeddings1, embeddings2)

# 필터링된 데이터셋에서 의미적 유사도 기반으로 추가 필터링
final_data = []
similarity_threshold = 0.0  # 코사인 유사도 기준

# 질문과 답변을 일괄 처리하기 위한 리스트
questions = [item['question'] for item in filtered_data]
answers = [item['answer'] for item in filtered_data]

# 질문과 답변을 배치 처리하여 임베딩 계산
question_embeddings = model.encode(questions, convert_to_tensor=True, batch_size=32)
answer_embeddings = model.encode(answers, convert_to_tensor=True, batch_size=32)

# 유사도 계산
similarities = cosine_similarity(question_embeddings, answer_embeddings)

# 유사도에 따라 데이터 필터링
for idx, similarity in enumerate(similarities):
    if similarity.item() >= similarity_threshold:
        final_data.append(filtered_data[idx])
    else:
        print(f"유사도 부족 - Question: {questions[idx][:100]}...")
        print(f"Answer: {answers[idx][:50]}...")
        print(f"유사도: {similarity.item()}")
        print("-" * 50)

# 최종 데이터셋 크기 확인
print(f"최종 필터링된 데이터셋 크기: {len(final_data)}")
```

```python
# 최종 필터링된 데이터를 새로운 JSON 파일로 저장
final_json_file_path = "/content/drive/MyDrive/final_filtered_라벨링데이터.json"

# JSON 파일로 저장
with open(final_json_file_path, 'w', encoding='utf-8') as f:
    json.dump(final_data, f, ensure_ascii=False, indent=4)

print(f"최종 필터링된 데이터셋이 저장되었습니다: {final_json_file_path}")
```

```python
import json
from sentence_transformers import SentenceTransformer, util

# 저장된 JSON 파일 로드
final_json_file_path = "/content/drive/MyDrive/final_filtered_라벨링데이터.json"
with open(final_json_file_path, 'r', encoding='utf-8') as f:
    final_data = json.load(f)

# Sentence-BERT 모델 로드
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# 사용자의 질문에 대해 가장 유사한 답변 찾기
def find_best_answer(user_question):
    # 사용자의 질문 임베딩
    user_question_embedding = model.encode(user_question, convert_to_tensor=True)

    best_answer = None
    best_similarity = -1  # 가장 높은 유사도를 찾기 위해 초기값을 음수로 설정

    for item in final_data:
        question = item['question']
        answer = item['answer']

        # 질문 임베딩 추출
        question_embedding = model.encode(question, convert_to_tensor=True)

        # 유사도 계산
        similarity = util.pytorch_cos_sim(user_question_embedding, question_embedding).item()

        # 가장 높은 유사도의 답변 선택
        if similarity > best_similarity:
            best_similarity = similarity
            best_answer = answer

    return best_answer, best_similarity

# 사용자 질문에 대한 테스트
user_question = "정책 5가지만 알려줘"
best_answer, best_similarity = find_best_answer(user_question)

print(f"질문: {user_question}")
print(f"가장 유사한 답변: {best_answer}")
print(f"유사도: {best_similarity}")
```

```python
import json
import re
from datasets import Dataset

# 1. JSON 파일 경로
json_file_path = "/content/drive/MyDrive/라벨링데이터2.json"

# 2. JSON 파일 읽기
with open(json_file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 3. 데이터가 잘 로드되었는지 확인
if len(data) == 0:
    print("데이터가 비어 있습니다. 파일을 확인해주세요.")
else:
    print(f"로드된 데이터 항목 수: {len(data)}")
    print(f"첫 번째 항목 예시: {data[0]}")

# 4. 데이터를 처리하여 필요한 정보 추출
processed_data = []
for item in data:
    context = item['question']  # 질문 대신에 문맥이라고 가정
    answer = item['answer']

    # 공백 및 불필요한 문자 제거
    context_clean = re.sub(r'\s+', ' ', context.strip())
    answer_clean = re.sub(r'\s+', ' ', answer.strip())

    # 답변이 문맥에서 어디에 있는지 찾습니다. (context에서 answer를 찾음)
    start_pos = context_clean.find(answer_clean)
    if start_pos == -1:
        print(f"문맥과 답변 매칭 오류: {item['question'][:30]}...")
        continue
    end_pos = start_pos + len(answer_clean)

    processed_data.append({
        "context": context_clean,  # 전체 문맥
        "question": answer_clean,  # 질문 대신 답변을 질문처럼 처리
        "start_positions": start_pos,  # 답변이 시작하는 위치
        "end_positions": end_pos  # 답변이 끝나는 위치
    })

# 5. 데이터셋 생성
if len(processed_data) == 0:
    print("처리된 데이터가 없습니다. 매칭 오류를 확인하세요.")
else:
    dataset = Dataset.from_dict({
        'context': [item['context'] for item in processed_data],
        'question': [item['question'] for item in processed_data],
        'start_positions': [item['start_positions'] for item in processed_data],
        'end_positions': [item['end_positions'] for item in processed_data]
    })

    print(f"데이터셋 크기: {len(dataset)}")
    print(f"첫 번째 데이터: {dataset[0]}")
```
